```python
mydictionary={'playerID':[],'yearID':[], 'teamID':[], 'lgID':[], 'salary':[]}
for row in df.iterrows():
    mydictionary['playerID'].append(row[0])
    mydictionary['yearID'].append(row[1].yearID)
    mydictionary['teamID'].append(row[1].teamID)
    mydictionary['lgID'].append(row[1].lgID)
    mydictionary['salary'].append(row[1].salary)
    
print(mydictionary)
```

```python
newdf=pd.DataFrame(data=mydictionary)
newdf
new_col= {'playerID': 'a', 'yearID': 'b', 'teamID': 'c', 'lgID':'d', 'salary':'e'}
newdf.rename(columns=new_col, inplace=True)
newdf
```

```python
x=np.linspace(-5,5,20)
y=np.array([j ** 2 for j in x])
plt.plot(x,y)
plt.legend(["parabola"],loc="best")
plt.title('PARABOLA')
plt.xlabel('x_label')
plt.ylabel('y_label')
plt.grid()
plt.show()
```

```python
x=np.linspace(-5,5,20)
np.random.shuffle(x)
y=np.array([j ** 2 for j in x])
plt.figure(figsize=(4,4))
plt.plot(x,y,color='green')
plt.legend(["random"],loc="best")
plt.title('PARABOLA')
plt.xlabel('x_label')
plt.ylabel('y_label')
plt.grid()
plt.show()
```

```python
x = np.linspace(-5, 5, num=20)
y1 = np.array([j for j in x])
y2 = np.array([j ** 2 for j in x])
y3 = np.array([j ** 3 for j in x])

plt.plot(x,y1,label="linear")
plt.legend(loc="best")
plt.text(-0.5,3,'$y=x$',fontsize=30)
plt.xlabel('XPOINTS')
plt.ylabel('YPOINTS')
plt.show()

plt.plot(x,y2,label="quadratic")
plt.legend(loc="best")
plt.text(-0.5,3,'$y=x^2$',fontsize=30)
plt.xlabel('XPOINTS')
plt.ylabel('YPOINTS')
plt.show()

plt.plot(x,y3,label="cubic")
plt.legend(loc="best")
plt.text(-0.5,3,'y=x^3',fontsize=30)
plt.xlabel('XPOINTS')
plt.ylabel('YPOINTS')
plt.show()
```

```python
plt.plot(x,y1, label='linear',color='cyan')
plt.plot(x,y2, label='quadratic',color='yellow')
plt.plot(x,y3, label='cubic',color='black')
plt.text(-1.5, 3, '$y=x$', fontsize=10,color='red')
plt.text(-3.5, 3, '$y=x^2$', fontsize=10)
plt.text(-5.5, 3, '$y=x^3$', fontsize=10)
plt.legend(loc='best')
plt.show()
```

```python
x = np.linspace(-5, 5, num=20)
y1 = np.array([j for j in x])
y2 = np.array([j ** 2 for j in x])
y3 = np.array([j ** 3 for j in x])

fig,axes=plt.subplots(nrows=3,ncols=1,figsize=(10,10))
print(fig)
print(axes)
fig.suptitle("Subplot plotting",fontsize=30)
axes[0].set_title("Sub plot 1: x vs y1",fontsize=9,fontstyle='oblique')
axes[0].plot(x,y1,color='red')

axes[1].plot(x,y2)
axes[1].set_title("Sub plot 1: x vs y2",fontsize=9,fontstyle='oblique')
axes[2].plot(x,y3,color='green')
axes[2].set_title("Sub plot 1: x vs y3",fontsize=9,fontstyle='oblique')
```
